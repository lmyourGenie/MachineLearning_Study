# 🍋1. Machine Learning

# 1.1 머신러닝이란 ?
* 컴퓨터가 학습할 수 있도록 하는 알고리즘과 기술을 개발하는 분야
* 더 많은 데이터의 축적을 통해 성능을 개선할 수 있도록 하는 다양한 알고리즘과 방법론
* __훈련 데이터를 통해 학습된 속성을 기반으로 한 예측__ 에 중점을 둔다.
<br/>

* _머신러닝의 목적_
    * _미래의 새로운 데이터를 더 정확하게 예측/추론하기 위함_

# 1.2 분류
<br/>

* 지도 학습 (Supervised Learning)
    * 회귀 (Regression)
    * 분류 (Classification)    
* 비지도 학습 (Unsupervised Learning)
    * 군집화 (Clustering)
    * 분포 추정 (Underlying Probability Density Estimation)

# 1.3 지도 학습 (Supervised Learning)
> 사람이 교사로써 각각의 입력 x에 대해 레이블 y를 달아 놓은 데이터를 컴퓨터에 입력해주는 방법

<br/>

### 1.3.1 회귀 (Regression)
- x를 쭉 뿌려놓고 이것을 가장 잘 설명하는 직선 하나 혹은 이차함수 곡선 하나를 그릴 때 회귀 기능을 사용한다.
시간에 따라 변화하는 데이터나 어떤 영향, 가설적 실험, 인과 관계의 모델링 등의 통계적 예측에 이용될 수 있다.  

예) 주가 예측

<br/>

### 1.3.2 분류 (Classification)
- 레이블 y가 이산적인 경우  즉, y가 가질 수 있는 값이 유한한 경우 분류 혹은 인식 문제라고 한다.  

예) 필기체 인식, 스팸 메일 분류

<br/>
<br/>

---


# 1.4 비지도 학습 (Unsupervised Learning)
> 사람 없이 컴퓨터가 스스로 레이블되어 있지 않은 데이터에 대해 학습하는 것

<br/>

### 1.4.1  군집화 (Clustering)
- 데이터만 쭉 뿌려져 있을 때, 즉 x만 사용하여 데이터 간 거리에 따라 군집을 학습하는 것  

예) 마케팅 고객 그룹화

<br/>

### 1.4.2  분포 추정 (Underlying Probability Density Estimation)
- 군집화에서 더 나아가서 x들이 쭉 뿌려져 있을 때 이 x들이 어떤 확률 분포에서 나온 샘플들인지 추정하는 것

<br/>
<br/>

---

# 1.5 강화 학습 (Reinforcement Learning)
> 현재 State에서 어떤 Action을 취하는 것이 최적인지 학습하는 것  
> 행동을 취할 때마다 외부 환경에서 __Reward__ 가 주어지며 이러한 보상을 최대화하는 방향으로 학습을 진행한다.

예) 알파고

# 🍋2. Deep Learning

# 2.1 딥러닝이란 ?

* 컴퓨터들이 인간의 두뇌와 비슷한 모양의 대형 `인공 신경망` 을 형성하는 일종의 기계 학습
* 사람의 개입이 필요한 기존의 지도 학습에 보다 능동적인 비 지도 학습이 결합돼, 마치 사람처럼 __스스로 학습__ 할 수 있는 인공지능 기술
<br/>
<br/>

---

# 2.2 딥러닝의 종류
## 2.2.1 DNN
` Deep Neural Network `

* 입력층(input layer)과 출력층(output layer) 사이에 여러 개의 은닉층(hidden layer)들로 이루어진 인공신경망(ANN)이다.  
* 일반적인 인공 신경망과 마찬가지로 복잡한 비선형 관계들을 모델링할 수 있다.
<br/>

## 2.2.2 CNN
` Convolutional Neural Network `

* 다계층 퍼셉트론(Multilayer perceptrons)의 한 종류이며, 하나 또는 여러 개의 합성곱 계층과 그 위에 올려진 일반적인 인공신경망 계층들로 이루어져 있다.  
* 가중치와 통합계층(pooling layer)들을 추가로 활용한다.
* 이미지를 인식하기위해 패턴을 찾는데 특히 유용하다.  

예) 자율주행자동차, 객체인식

<br/>

## 2.2.3 RNN
` Recurrent Neural Network `

* 인공신경망을 구성하는 유닛 사이의 연결이 Directed cycle을 구성하는 신경망을 말한다.
* 장점 : 반복적이고 순차적인 데이터 학습에 특화 
    * 예) 시간의 흐름에 따라 변화하는 Data 학습용
* 단점 : 처음 시작하는 weight의 값이 학습이 점차 진행될수록 상쇄된다.
    * 이를 보완한 Network : LSTM
